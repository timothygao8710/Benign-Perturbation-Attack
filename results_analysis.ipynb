# Micro

```python
f = open('./big_one_percentage_fixed.json')
mp = json.load(f)
```

```python
print(mp.keys())
print(len(mp.keys()))
```

```python
for i in mp.keys():
    print(i)
    print(mp[i]['WA'])
```

```python
tot_WAs = 0
tot_CAs = 0

dist_percentages = []

for i in mp.keys():
    tot_WAs += mp[i]['WA']
    tot_CAs += mp[i]['CA']
    dist_percentages.append(mp[i]['WA']/(mp[i]['WA']+mp[i]['CA']))
```

```python
plt.hist(dist_percentages, bins=50)
```

```python
print('Mean:', sum(dist_percentages)/len(dist_percentages))
print('Max:', max(dist_percentages))
print('Min:', min(dist_percentages))
print('Std:', np.std(dist_percentages))
print('Median:', np.median(dist_percentages))

print('Percentage of non-zero:', len([i for i in dist_percentages if i != 0])/len(dist_percentages))
```

# Macro

```python
combined_mp = {}
percentages = [0.1, 0.01, 0.2, 0.02, 0.3, 0.03, 0.5, 0.05, 0.07, 0.12, 0.15, 0.75, 1]

for i in percentages:
    f = open('./' + str(i) + ' ' + 'out_percentage.json')
    print(f)
    mp = json.load(f)
    
    print(mp.keys())
    print(type(mp))
    
    combined_mp[i] = mp
```

```python
with open('./all_out_percentages.json', 'w') as out_file:
    json.dump(combined_mp, out_file)
```

```python
x = []
y = []

for p in combined_mp:
    x.append(p)
    
    mp = combined_mp[p]
    
    tot_WAs = 0
    tot_CAs = 0

    dist_percentages = []

    for i in mp.keys():
        tot_WAs += mp[i]['WA']
        tot_CAs += mp[i]['CA']
    
    y.append(tot_CAs / (tot_WAs + tot_CAs))
```

```python
paired = list(zip(x, y))

# Sort the pairs according to the first element (elements in x)
paired_sorted = sorted(paired, key=lambda pair: pair[0])

# Unzip the pairs back into two separate arrays
x_sorted, y_sorted = zip(*paired_sorted)

# Convert the tuples back into lists (optional)
x = list(x_sorted)
y = list(y_sorted)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', marker='o', linestyle='-')
plt.title('Line Chart of [x, y] Coordinates')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
