```python
import pandas as pd
from string_functions import *
from tqdm import tqdm
import json
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('./out.json')
mp = json.load(f)
```

```python
print(mp.keys())
print(len(mp.keys()))
```

```python
for i in mp.keys():
    print(i)
    print(mp[i]['WA'])
```

```python
tot_WAs = 0
tot_CAs = 0

dist_percentages = []

for i in mp.keys():
    tot_WAs += len(mp[i]['WA'])
    tot_CAs += len(mp[i]['CA'])
    dist_percentages.append(len(mp[i]['WA'])/(len(mp[i]['WA'])+len(mp[i]['CA'])))
```

```python
plt.hist(dist_percentages, bins=50)
```

```python
print('Mean:', sum(dist_percentages)/len(dist_percentages))
print('Max:', max(dist_percentages))
print('Min:', min(dist_percentages))
print('Std:', np.std(dist_percentages))
print('Median:', np.median(dist_percentages))

print('Percentage of non-zero:', len([i for i in dist_percentages if i != 0])/len(dist_percentages))
```
